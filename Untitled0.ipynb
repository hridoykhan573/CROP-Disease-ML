```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
import os

# --- AUTO-PATH FINDER ---
# Dataset folder-ti kothay ache sheta auto khuje ber korbe
def find_path(target_folder):
    for root, dirs, files in os.walk("."):
        if target_folder in dirs:
            return os.path.join(root, target_folder)
    return None

# 'Train' folder-ti khujchi
TRAIN_PATH = find_path("Train")

if TRAIN_PATH is None:
    print("[ERROR] 'Train' folder-ti pawa jani. 'archive.zip' unzip hoyeche ki na check koro.")
else:
    print(f"[INFO] Dataset found at: {TRAIN_PATH}")

    # --- STAGE 2: DATASET LOADING ---
    IMAGE_SIZE = 256
    BATCH_SIZE = 32

    # Loading dataset (Requirement 2 from Project Workflow)
    try:
        train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
            TRAIN_PATH,
            shuffle=True,
            image_size=(IMAGE_SIZE, IMAGE_SIZE),
            batch_size=BATCH_SIZE
        )

        class_names = train_dataset.class_names
        print(f"Classes: {class_names}")

        # --- STAGE 3: RAW BASELINE MODEL ---
        # Requirement: Minimal preprocessing to run the code
        model = models.Sequential([
            # Pixel scaling 0 to 1
            layers.Rescaling(1./255, input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)),

            # Simple CNN Architecture
            layers.Conv2D(32, (3, 3), activation='relu'),
            layers.MaxPooling2D((2, 2)),

            layers.Conv2D(64, (3, 3), activation='relu'),
            layers.MaxPooling2D((2, 2)),

            layers.Flatten(),
            layers.Dense(64, activation='relu'),
            layers.Dense(len(class_names), activation='softmax')
        ])

        # Compiling the baseline model
        model.compile(
            optimizer='adam',
            loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
            metrics=['accuracy']
        )

        # Baseline Training (Expected output: Baseline results)
        print("\nStarting Baseline Training (5 Epochs)...")
        history = model.fit(train_dataset, epochs=5)

        print("\n[SUCCESS] Stage 3 complete. Model is trained on Raw Data.")

    except Exception as e:
        print(f"[ERROR] Loading failed: {e}")
```

```python
# ==========================================
# STAGE 4: PREPROCESSING (AUGMENTATION)
# ==========================================
# Requirement: Handle scaling, encoding, and image augmentation

# Defining Augmentation Layer
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.1),
    layers.RandomBrightness(0.1),
])

# ==========================================
# STAGE 5: CLEANED-DATA MODELS
# ==========================================
# Requirement: Train the same model after preprocessing to compare

# Building the Improved Model with Augmentation
improved_model = models.Sequential([
    # Input Layer (Warning avoid korar jonno)
    layers.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3)),

    # Applying Preprocessing/Augmentation (Stage 4)
    data_augmentation,

    # Scaling pixel values
    layers.Rescaling(1./255),

    # CNN Architecture (Same as Baseline for fair comparison)
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(len(class_names), activation='softmax')
])

improved_model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)

print("\n[INFO] Starting Training with Augmented/Cleaned Data...")
history_improved = improved_model.fit(train_dataset, epochs=5)

# ==========================================
# COMPARISON VISUALIZATION (Stage 5 Requirement)
# ==========================================
plt.figure(figsize=(12, 5))

# Plotting Accuracy Comparison
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Baseline (Raw)')
plt.plot(history_improved.history['accuracy'], label='Improved (Augmented)')
plt.title('Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plotting Loss Comparison
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Baseline (Raw)')
plt.plot(history_improved.history['loss'], label='Improved (Augmented)')
plt.title('Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()

print("\n[SUCCESS] Stage 4 & 5 Complete. Comparison graph generated.")
```

```python
# ==========================================
# STAGE 6: NEURAL NETWORK (TRANSFER LEARNING)
# ==========================================
# Requirement: Train MobileNetV2 or ResNet50 for high performance

print("\n[INFO] Initializing Transfer Learning with MobileNetV2...")

# 1. Base model loading (pre-trained on ImageNet)
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False # Feature extraction mode

# 2. Building the Transfer Learning Model
nn_model = models.Sequential([
    layers.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3)),
    data_augmentation, # Stage 4 augmentation included
    layers.Rescaling(1./255),
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(len(class_names), activation='softmax')
])

nn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print("\n[INFO] Training Neural Network (Stage 6)...")
history_nn = nn_model.fit(train_dataset, epochs=5)

# ==========================================
# STAGE 7: ENSEMBLE LEARNING (SOFT VOTING)
# ==========================================
# Requirement: Combine models using voting or averaging

print("\n[INFO] Creating Ensemble Model (Soft Voting)...")

def ensemble_predict(images):
    # Predict with both Improved CNN and MobileNetV2
    p1 = improved_model.predict(images)
    p2 = nn_model.predict(images)

    # Guidelines: Averaging predictions from 2 models (Soft Voting)
    final_prediction = (p1 + p2) / 2
    return final_prediction

# Testing Ensemble on one batch
for images, labels in train_dataset.take(1):
    preds = ensemble_predict(images)
    final_classes = np.argmax(preds, axis=1)

    # Accuracy check for ensemble on this batch
    batch_acc = np.mean(final_classes == labels.numpy())
    print(f"\n[SUCCESS] Ensemble Batch Accuracy: {batch_acc * 100:.2f}%")
    break

# ==========================================
# FINAL EVALUATION: CONFUSION MATRIX (Requirement)
# ==========================================
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Batch predict for Confusion Matrix
for images, labels in train_dataset.take(1):
    y_pred = np.argmax(nn_model.predict(images), axis=1)
    cm = confusion_matrix(labels, y_pred)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix - Neural Network (Stage 6)')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.show()
    break

print("\n[INFO] Stage 6 & 7 Complete. Final Ensemble result and Confusion Matrix ready.")
```

```python
# ==========================================
# FINAL DEMO: PREDICTION ON TEST SAMPLES
# ==========================================
import numpy as np

# Ekta sample batch nawa dataset theke
for images, labels in train_dataset.take(1):
    # Ensemble method use kore prediction kora
    preds = ensemble_predict(images)

    plt.figure(figsize=(15, 8))
    # Prothom 3ti image-er jonno prediction dekhalchi
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(images[i].numpy().astype("uint8"))

        actual_class = class_names[labels[i]]
        predicted_class = class_names[np.argmax(preds[i])]
        confidence = np.max(preds[i]) * 100

        # Title e color coding (Correct hole Shobuj, Vul hole Lal)
        title_color = "green" if actual_class == predicted_class else "red"

        plt.title(f"Actual: {actual_class}\nPred: {predicted_class}\nConf: {confidence:.2f}%",
                  color=title_color, fontsize=12)
        plt.axis("off")
    plt.show()
    break
```
